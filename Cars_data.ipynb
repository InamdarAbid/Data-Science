# Data Wrangling

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv',header = None,names = ["buying","maint","doors","persons","lug_boot","safety","acceptability"])
df.head()
```

```python
df.dtypes
```

```python
df['buying_price'] = df.buying.astype("category").cat.codes
```

```python
df
```

```python
df.head()
```

```python
df['maintenance'] = df.maint.astype("category",ordered = True).cat.codes
df['luggage_boot'] = df.lug_boot.astype("category",ordered = True).cat.codes
df["safety1"] = df.safety.astype("category",ordered = True).cat.codes
df["accept"] = df.acceptability("category",ordered = True).cat.codes
df.head()
```

```python
df.head()
```

```python
df.drop(columns = ["buying","maint","lug_boot","safety","acceptability"],inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['doors2'] = pd.to_numeric(df.doors, errors = 'coerce')
df['persons2'] = pd.to_numeric(df.persons, errors = 'coerce')
df.head()
```

```python
df.dtypes
```

```python
df.drop(columns = ["doors","persons"], inplace = True)
```

```python
df.head()
```

```python
df.mean(axis = 0)
```

```python
df.fillna(df.mean(axis = 0,), inplace = True)
df.isna()
```
